### **Jarque-Bera**

## Pengertian

Uji Jarque-Bera adalah metode statistik yang digunakan untuk
mengetahui apakah data berdistribusi normal atau tidak.

Uji ini menggunakan dua karakteristik distribusi data yaitu:

1. Skewness (kemencengan data)
2. Kurtosis (keruncingan data)

Apabila nilai skewness mendekati 0 dan kurtosis mendekati 3,
maka data cenderung berdistribusi normal.

### Langkah-Langkah Perhitungan

1. Menentukan data yang akan diuji normalitasnya.

2. Menghitung jumlah data (n).

3. Menghitung nilai rata-rata (mean).

4. Menghitung selisih setiap data terhadap rata-rata.

5. Menghitung momen ke-2 (m₂).

6. Menghitung momen ke-3 (m₃).

7. Menghitung momen ke-4 (m₄).

8. Menghitung nilai skewness menggunakan m₂ dan m₃.

9. Menghitung nilai kurtosis menggunakan m₂ dan m₄.

10. Menghitung nilai statistik Jarque-Bera (JB).

11. Membandingkan nilai JB dengan nilai kritis Chi-Square.

12. Menentukan keputusan dan menarik kesimpulan.

### Kelebihan Uji Jarque-Bera
1. Mudah dihitung dan dipahami.
2. Menggunakan dua karakteristik distribusi sekaligus, yaitu skewness dan kurtosis.
3. Cocok digunakan untuk menguji asumsi normalitas pada data statistik maupun residual regresi.
4. Tidak memerlukan visualisasi grafik untuk menentukan normalitas data.

### Kekurangan Uji Jarque-Bera
1. Sensitif terhadap ukuran sampel yang kecil.
2. Kurang akurat jika jumlah data sangat sedikit.
3. Hanya berfokus pada skewness dan kurtosis sehingga tidak selalu mendeteksi seluruh bentuk penyimpangan distribusi normal.
4. Dapat dipengaruhi oleh keberadaan outlier pada data.

### Rumus Jarque-Bera

$$
JB=
\frac{n}{6}
\left(
S^2+
\frac{(K-3)^2}{4}
\right)
$$

Keterangan:
$JB$ = Statistik Jarque-Bera

$n$ = Jumlah data (observasi)

$S$ = Skewness (kemencengan data)

$K$ = Kurtosis (keruncingan data)

Rumus Skewness:
$$
S=
\frac{m_3}
{m_2^{3/2}}
$$

Keterangan:

$m_2$ = momen ke-2

$m_3$ = momen ke-3

Rumus Kurtosis:

$$
K=
\frac{m_4}
{m_2^2}
$$

Keterangan:

$m_4$ = momen ke-4

```python
data = [12,15,14,10,18,16,13,17,11,14]

n = len(data)

mean = sum(data) / n

selisih = []
for x in data:
    selisih.append(x - mean)

m2 = 0
for d in selisih:
    m2 += d**2
m2 = m2 / n

m3 = 0
for d in selisih:
    m3 += d**3
m3 = m3 / n

m4 = 0
for d in selisih:
  m4 += d**4
m4 = m4 / n

skewness = m3 / (m2**1.5)
kurtosis = m4 / (m2**2)

jb = (n / 6) * (skewness**2 + ((kurtosis - 3)**2) / 4)

chi_square_critical = 5.991

if jb > chi_square_critical:
    keputusan = "Tolak H0: data tidak berdistribusi normal"
else:
    keputusan = "Gagal menolak H0: data berdistribusi normal"

print(f"JB Statistic: {jb:.4f}")
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")
print(f"Decision: {keputusan}")
```

**Jarque-Bera (Test)**

```python
def jarque_bera_test(data):

    # jumlah data
    n = len(data)

    # mean
    mean = sum(data) / n

    # hitung selisih
    selisih = []
    for x in data:
        selisih.append(x - mean)

    # momen ke-2
    m2 = 0
    for d in selisih:
        m2 += d**2
    m2 = m2 / n

    # momen ke-3
    m3 = 0
    for d in selisih:
        m3 += d**3
    m3 = m3 / n

    # momen ke-4
    m4 = 0
    for d in selisih:
        m4 += d**4
    m4 = m4 / n

    # skewness
    skewness = m3 / (m2 ** 1.5)

    # kurtosis
    kurtosis = m4 / (m2 ** 2)

    # Jarque-Bera
    jb = (n / 6) * (skewness**2 + ((kurtosis - 3)**2) / 4)

    # keputusan
    chi_square_critical = 5.991

    if jb > chi_square_critical:
        keputusan = "Tolak H0: data tidak berdistribusi normal"
    else:
        keputusan = "Gagal menolak H0: data berdistribusi normal"

    # output seluruh nilai
    print("n =", n)
    print("mean =", mean)
    print("m2 =", m2)
    print("m3 =", m3)
    print("m4 =", m4)
    print("skewness =", skewness)
    print("kurtosis =", kurtosis)
    print("jb =", jb)
    print(keputusan)

data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]
jarque_bera_test(data)
```

### Kesimpulan
Berdasarkan hasil Uji Jarque-Bera diperoleh nilai statistik Jarque-Bera (JB) sebesar 0,4449. Nilai tersebut lebih kecil daripada nilai kritis Chi-Square sebesar 5,991 sehingga keputusan yang diperoleh adalah gagal menolak H₀. Dengan demikian dapat disimpulkan bahwa data berdistribusi normal. Hasil ini menunjukkan bahwa data telah memenuhi asumsi normalitas sehingga dapat digunakan pada analisis statistik parametrik yang mensyaratkan distribusi normal.

### **Uji Normalitas Kolmogorov-Smirnov**

## Pengertian

Uji Kolmogorov-Smirnov (K-S) adalah metode statistik yang digunakan
untuk menguji apakah distribusi data sampel mengikuti distribusi
teoritis tertentu, biasanya distribusi normal.

Prinsip dasar uji ini adalah membandingkan distribusi kumulatif
empiris data dengan distribusi kumulatif teoritis. Semakin kecil
perbedaannya, semakin besar kemungkinan data mengikuti distribusi
yang diuji.

## Hipotesis

H0 : Data berdistribusi normal

H1 : Data tidak berdistribusi normal

## Rumus

$$
D = \max |F_o(x)-F_t(x)|
$$

Keterangan:

$D$ = Statistik Kolmogorov-Smirnov

$\max$ = Nilai maksimum

$F_o(x)$ = Distribusi kumulatif empiris (Empirical Cumulative

Distribution Function/ECDF)

$F_t(x)$ = Distribusi kumulatif teoritis (CDF distribusi normal)

$x$ = Nilai observasi data

## Interpretasi Rumus

Uji Kolmogorov-Smirnov menghitung selisih absolut antara
distribusi kumulatif empiris dan distribusi kumulatif teoritis
pada setiap titik data.

Nilai selisih terbesar dipilih sebagai nilai statistik uji D.

Semakin kecil nilai D, maka data semakin mendekati distribusi
normal.

### Langkah-Langkah Logika Pemrograman

1. Menyiapkan data.
2. Mengurutkan data dari kecil ke besar.
3. Menghitung rata-rata (mean).
4. Menghitung standar deviasi.
5. Menghitung nilai Z-Score.
6. Menghitung distribusi kumulatif teoritis (CDF).
7. Menghitung distribusi kumulatif empiris (ECDF).
8. Menghitung selisih absolut keduanya.
9. Mengambil nilai maksimum sebagai statistik D.
10. Membandingkan D hitung dengan D kritis.
11. Menentukan keputusan pengujian.

### Kelebihan Uji Kolmogorov-Smirnov

1. Dapat digunakan untuk membandingkan distribusi data sampel dengan distribusi teoritis tertentu, seperti distribusi normal.
2. Tidak hanya memperhatikan kemencengan (skewness) atau keruncingan (kurtosis), tetapi seluruh bentuk distribusi data.
3. Mudah diterapkan pada berbagai jenis data kontinu.
4. Memberikan hasil yang objektif melalui perbandingan statistik D hitung dan D kritis.

### Kekurangan Uji Kolmogorov-Smirnov
1. Kurang sensitif dalam mendeteksi penyimpangan distribusi pada ukuran sampel yang kecil.
2. Hasil pengujian dapat dipengaruhi oleh jumlah sampel yang digunakan.
3. Lebih sensitif terhadap perbedaan di bagian tengah distribusi dibandingkan pada bagian ekor distribusi.
4. Perhitungannya relatif lebih kompleks dibandingkan beberapa uji normalitas lainnya seperti Jarque-Bera.

```python
import math

def kolmogorov_smirnov_test(data):

    n = len(data)

    # Urutkan Data
    sorted_data = sorted(data)

    # Menghitung Mean
    mean = sum(data) / n

    # Hitung Standar Deviasi
    variance = 0
    for x in data:
        variance += (x - mean)**2
    variance = variance / n
    std = variance**0.5

    D = 0.0

    D_kritis = 1.36 / math.sqrt(n)

    for i, x_val in enumerate(sorted_data, start=1):

        # Hitung Z-Score
        z = (x_val - mean) / std

        # Hitung CDF Normal
        cdf = 0.5 * (1 + math.erf(z / math.sqrt(2)))

        # Hitung ECDF
        ecdf = i / n

        # Hitung selisih absolut
        diff = abs(ecdf - cdf)

        # Cari nilai D maksimum
        D = max(D, diff)

    print("D Hitung =", round(D, 4))
    print("D Kritis =", round(D_kritis, 4))

    if D < D_kritis:
        keputusan = "Gagal menolak H0: data berdistribusi normal"
    else:
        keputusan = "Tolak H0: data tidak berdistribusi normal"
    print(keputusan)

data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]
kolmogorov_smirnov_test(data)
```

### Kesimpulan
Berdasarkan hasil Uji Kolmogorov-Smirnov diperoleh nilai D hitung sebesar 0,1000 dan nilai D kritis sebesar 0,4301. Karena nilai D hitung lebih kecil daripada nilai D kritis, maka keputusan yang diperoleh adalah gagal menolak H₀. Dengan demikian dapat disimpulkan bahwa data berdistribusi normal. Hasil ini menunjukkan bahwa distribusi data sampel tidak berbeda secara signifikan dengan distribusi normal teoritis sehingga asumsi normalitas telah terpenuhi.

### **Uji Homogenitas F Varians**

## Pengertian

Uji F Varians digunakan untuk mengetahui apakah dua kelompok data
memiliki varians yang sama (homogen) atau berbeda secara signifikan.

Uji ini dilakukan dengan membandingkan varians terbesar terhadap
varians terkecil.

### Rumus
$$
F=
\frac{s_1^2}
{s_2^2}
$$

Keterangan:

$F$ = Statistik Uji F

$s_1^2$ = Varians yang lebih besar

$s_2^2$ = Varians yang lebih kecil

### Langkah-Langkah Perhitungan

1. Menyiapkan dua kelompok data yang akan dibandingkan.
2. Menghitung jumlah data masing-masing kelompok (n₁ dan n₂).
3. Menghitung nilai rata-rata (mean) setiap kelompok data.
4. Menghitung jumlah kuadrat selisih setiap data terhadap rata-ratanya.
5. Menghitung varians kelompok pertama.
6. Menghitung varians kelompok kedua.
7. Menentukan varians terbesar dan varians terkecil.
8. Menghitung nilai F hitung dengan membagi varians terbesar terhadap varians terkecil.
9. Menentukan nilai F tabel berdasarkan taraf signifikansi dan derajat bebas.
10. Membandingkan nilai F hitung dengan F tabel.
11. Menentukan keputusan pengujian dan menarik kesimpulan.

### Kelebihan Uji F Varians
1. Mudah dipahami dan dihitung karena hanya menggunakan nilai varians kedua kelompok.
2. Efektif untuk menguji kesamaan varians antara dua kelompok data.
3. Sering digunakan sebagai syarat awal sebelum melakukan uji parametrik seperti uji t dan ANOVA.
4. Memberikan keputusan yang jelas mengenai homogenitas varians.

### Kekurangan Uji F Varians
1. Sangat sensitif terhadap pelanggaran asumsi normalitas data.
2. Hasil pengujian dapat dipengaruhi oleh adanya nilai ekstrem (outlier).
3. Hanya dapat digunakan untuk membandingkan dua kelompok varians sekaligus.
4. Kurang robust dibandingkan uji homogenitas lainnya seperti Uji Levene apabila data tidak berdistribusi normal.

```python
def uji_F_varians(data1,data2):

    n1 = len(data1)
    n2 = len(data2)

    mean1 = sum(data1)/n1
    mean2 = sum(data2)/n2

    jumlah_kuadrat1 = 0
    for x in data1:
        jumlah_kuadrat1 += (x-mean1)**2

    varians1 = jumlah_kuadrat1/(n1-1)

    jumlah_kuadrat2 = 0
    for x in data2:
        jumlah_kuadrat2 += (x-mean2)**2

    varians2 = jumlah_kuadrat2/(n2-1)

    if varians1 > varians2:
        f_hitung = varians1/varians2
    else:
        f_hitung = varians2/varians1

    f_tabel = 6.39

    if f_hitung > f_tabel:
        keputusan = "Tolak H0: varians tidak homogen"
    else:
        keputusan = "Gagal menolak H0: varians homogen"

    print("n1 =", n1)
    print("n2 =", n2)
    print("mean1 =", mean1)
    print("mean2 =", mean2)
    print("varians1 =", varians1)
    print("varians2 =", varians2)
    print("f_hitung =", f_hitung)
    print("f_tabel =", f_tabel)
    print("keputusan =", keputusan)

data1 = [70,75,80,72,78]
data2 = [65,85,60,90,70]

uji_F_varians(data1,data2)
```

### Kesimpulan
Berdasarkan hasil Uji Homogenitas F Varians diperoleh nilai F hitung sebesar 9,8529 dan nilai F tabel sebesar 6,39. Karena nilai F hitung lebih besar daripada nilai F tabel, maka keputusan yang diperoleh adalah menolak H₀. Dengan demikian dapat disimpulkan bahwa kedua kelompok data memiliki varians yang tidak homogen. Hasil ini menunjukkan bahwa tingkat penyebaran data pada kedua kelompok berbeda secara signifikan, sehingga asumsi homogenitas varians tidak terpenuhi.

### **Uji Levene**

### Pengertian

Uji Levene digunakan untuk menguji kesamaan varians dari dua atau lebih kelompok data.

Keunggulan Levene dibanding Uji F adalah:

1. Lebih robust terhadap data yang tidak normal.
2. Sering digunakan sebelum ANOVA.

Hipotesis:

1. H₀ : Varians seluruh kelompok sama (homogen).
2. H₁ : Minimal ada satu varians yang berbeda.

### Rumus
$$
Z_{ij}=|X_{ij}-\bar{X_i}|
$$

Keterangan:

$Z_{ij}$ = deviasi absolut data ke-j pada kelompok ke-i

$X_{ij}$ = nilai data ke-j pada kelompok ke-i

$\bar{X_i}$ = rata-rata kelompok ke-i

$i$ = kelompok data

$j$ = urutan data dalam kelompok

## Langkah-Langkah Logika Pemrograman Uji Levene

1. Menyiapkan dua kelompok data yang akan diuji homogenitas variansnya.

2. Menghitung rata-rata masing-masing kelompok.

3. Menghitung deviasi absolut setiap data terhadap rata-rata kelompoknya dengan rumus:

$$
Z_{ij}=|X_{ij}-\bar X_i|
$$

4. Menyimpan seluruh nilai deviasi absolut ke dalam kelompok baru.

5. Menghitung rata-rata dari setiap kelompok deviasi absolut.

6. Menghitung rata-rata keseluruhan deviasi absolut.

7. Menghitung variasi antar kelompok (SS Between).

8. Menghitung variasi dalam kelompok (SS Within).

9. Menghitung Mean Square Between (MS Between).

10. Menghitung Mean Square Within (MS Within).

11. Menghitung statistik Levene:

$$
W=
\frac{MS_{Between}}
{MS_{Within}}
$$

12. Membandingkan nilai statistik Levene dengan nilai F tabel.

13. Membuat keputusan pengujian.

### Kelebihan Uji Levene

1. Lebih robust terhadap pelanggaran asumsi normalitas dibandingkan Uji F Varians.
2. Dapat digunakan untuk dua kelompok atau lebih sekaligus.
3. Tidak terlalu sensitif terhadap data yang tidak berdistribusi normal.
4. Sering digunakan sebagai uji prasyarat sebelum melakukan ANOVA.
5. Memberikan hasil yang lebih stabil ketika terdapat perbedaan bentuk distribusi data antar kelompok.

### Kekurangan Uji Levene

1. Perhitungan lebih kompleks dibandingkan Uji F Varians.
2. Membutuhkan beberapa tahapan tambahan seperti perhitungan deviasi absolut dan variasi antar kelompok.
3. Pada ukuran sampel yang sangat kecil, hasil pengujian dapat menjadi kurang sensitif.
4. Masih dapat dipengaruhi oleh keberadaan outlier yang ekstrem.
5. Membutuhkan interpretasi yang lebih panjang dibandingkan uji homogenitas sederhana.

```python
def median(data):

    data = sorted(data)
    n = len(data)

    if n % 2 == 0:
        return (data[n//2-1] + data[n//2])/2
    else:
        return data[n//2]


data1 = [70,75,80,72,78]
data2 = [65,85,60,90,70]

median1 = median(data1)
median2 = median(data2)

z1 = []
for x in data1:
    z1.append(abs(x-median1))

z2 = []
for x in data2:
    z2.append(abs(x-median2))

mean_z1 = sum(z1)/len(z1)
mean_z2 = sum(z2)/len(z2)

grand_mean = (sum(z1)+sum(z2))/(len(z1)+len(z2))

ss_between = (
len(z1)*(mean_z1-grand_mean)**2 +
len(z2)*(mean_z2-grand_mean)**2
)

ss_within = 0

for z in z1:
    ss_within += (z-mean_z1)**2

for z in z2:
    ss_within += (z-mean_z2)**2

df_between = 1
df_within = len(z1)+len(z2)-2

f_hitung = (ss_between/df_between)/(ss_within/df_within)

f_tabel = 5.32

if f_hitung > f_tabel:
    keputusan = "Tolak H0: varians tidak homogen"
else:
    keputusan = "Gagal menolak H0: varians homogen"

print("median1 =", median1)
print("median2 =", median2)
print("mean_z1 =", mean_z1)
print("mean_z2 =", mean_z2)
print("ss_between =", ss_between)
print("ss_within =", ss_within)
print("f_hitung =", f_hitung)
print("f_tabel =", f_tabel)
print("keputusan =", keputusan)
```

### Kesimpulan
Berdasarkan hasil Uji Levene diperoleh nilai F hitung sebesar 3,4663 dan nilai F tabel sebesar 5,32. Karena nilai F hitung lebih kecil daripada nilai F tabel, maka keputusan yang diperoleh adalah gagal menolak H₀. Dengan demikian dapat disimpulkan bahwa kedua kelompok data memiliki varians yang homogen. Hasil ini menunjukkan bahwa tingkat penyebaran data pada kedua kelompok relatif sama sehingga asumsi homogenitas varians terpenuhi dan data layak digunakan untuk analisis statistik yang mensyaratkan homogenitas varians.

### **Uji Independensi Chi-Square**

## Pengertian

Uji Independensi Chi-Square digunakan untuk mengetahui
apakah terdapat hubungan antara dua variabel kategorik.

Jika nilai Chi-Square hitung lebih besar dari Chi-Square tabel,
maka kedua variabel dianggap memiliki hubungan.

### Rumus

$$
E_{ij}
=
\frac
{(Total\ Baris_i)(Total\ Kolom_j)}
{Grand\ Total}
$$

$$
\chi^2
=
\sum
\frac{(O_{ij}-E_{ij})^2}
{E_{ij}}
$$

Keterangan:
$O_{ij}$ = Frekuensi observasi

$E_{ij}$ = Frekuensi harapan

$i$ = Baris ke-i

$j$ = Kolom ke-j

### Langkah-Langkah Logika Pemrograman Uji Chi-Square

1. Menyiapkan data observasi dalam bentuk tabel kontingensi.
2. Menentukan jumlah baris dan jumlah kolom pada tabel.
3. Menghitung total setiap baris.
4. Menghitung total setiap kolom.
5. Menghitung grand total (jumlah seluruh data).
6. Menghitung frekuensi harapan (Expected Frequency) untuk setiap sel menggunakan total baris, total kolom, dan grand total.
7. Menghitung nilai Chi-Square pada setiap sel menggunakan selisih antara frekuensi observasi dan frekuensi harapan.
8. Menjumlahkan seluruh nilai Chi-Square untuk memperoleh nilai $\chi^2$ hitung.
9. Menentukan nilai $\chi^2$ tabel berdasarkan derajat kebebasan dan taraf signifikansi yang digunakan.
10. Membandingkan nilai $\chi^2$ hitung dengan $\chi^2$ tabel.
Menentukan keputusan pengujian.
11. Menarik kesimpulan mengenai ada atau tidaknya hubungan antar variabel.

### Kelebihan
1. Dapat digunakan untuk mengetahui hubungan antara dua variabel kategorik.
2. Tidak memerlukan asumsi distribusi normal pada data.
3. Mudah dipahami dan diterapkan pada berbagai bidang penelitian.
4. Dapat digunakan pada tabel kontingensi dengan ukuran yang beragam.
5. Perhitungannya relatif sederhana dan mudah diimplementasikan dalam program.

### Kekurangan
1. Hanya dapat digunakan untuk data kategorik atau data berbentuk frekuensi.
2. Tidak menunjukkan seberapa kuat hubungan antar variabel, hanya menunjukkan ada atau tidaknya hubungan.
3. Hasil pengujian dapat menjadi kurang akurat apabila frekuensi harapan pada beberapa sel terlalu kecil.
4. Sensitif terhadap ukuran sampel yang sangat besar sehingga perbedaan kecil dapat terlihat signifikan.
5. Tidak dapat digunakan secara langsung pada data numerik kontinu tanpa proses kategorisasi terlebih dahulu.

```python
def uji_chi_square(observasi):

    jumlah_baris = len(observasi)
    jumlah_kolom = len(observasi[0])

    total_baris = []
    for baris in observasi:
        total_baris.append(sum(baris))

    total_kolom = []

    for j in range(jumlah_kolom):

        total = 0

        for i in range(jumlah_baris):
            total += observasi[i][j]

        total_kolom.append(total)

    grand_total = sum(total_baris)

    harapan = []

    for i in range(jumlah_baris):

        baris = []

        for j in range(jumlah_kolom):

            eij = (total_baris[i]*total_kolom[j])/grand_total

            baris.append(eij)

        harapan.append(baris)

    chi_square = 0

    for i in range(jumlah_baris):
        for j in range(jumlah_kolom):

            O = observasi[i][j]
            E = harapan[i][j]

            chi_square += ((O-E)**2)/E

    chi_tabel = 3.841

    if chi_square > chi_tabel:
        keputusan = "Tolak H0: terdapat hubungan"
    else:
        keputusan = "Gagal menolak H0: tidak terdapat hubungan"

    print("jumlah_baris =", jumlah_baris)
    print("jumlah_kolom =", jumlah_kolom)
    print("grand_total =", grand_total)
    print("harapan =", harapan)
    print("chi_square =", chi_square)
    print("chi_tabel =", chi_tabel)
    print("keputusan =", keputusan)

data = [[30,20],[15,35]]

uji_chi_square(data)
```

### Kesimpulan

Berdasarkan hasil Uji Independensi Chi-Square diperoleh nilai $\chi^2$ hitung sebesar **9,0909** dan nilai $\chi^2$ tabel sebesar **3,841**. Karena nilai $\chi^2$ hitung lebih besar daripada nilai $\chi^2$ tabel $(9,0909 > 3,841)$, maka $H_0$ ditolak. Dengan demikian, dapat disimpulkan bahwa terdapat hubungan yang signifikan antara kedua variabel kategorik yang diuji, sehingga kedua variabel tersebut tidak bersifat independen.

### **Fisher's Exact Test**

**Pengertian**

Fisher's Exact Test adalah uji statistik yang digunakan untuk menguji hubungan (independensi) antara dua variabel kategorik pada tabel kontingensi 2×2.

Uji ini sangat cocok digunakan apabila:

1. ukuran sampel kecil.
2. terdapat frekuensi harapan kurang dari 5.
3. hasil Chi-Square kurang akurat.

### Rumus
$$
P=
\frac
{
(a+b)!(c+d)!(a+c)!(b+d)!
}
{
a!b!c!d!n!
}
$$

Keterangan:

$a,b,c,d$ = Frekuensi pada setiap sel tabel 2×2

$n$ = Total seluruh observasi

$P$ = Probabilitas Fisher Exact Test

## Langkah-Langkah Logika Pemrograman Fisher's Exact Test

1. Menyiapkan data dalam bentuk tabel kontingensi 2×2.

2. Menentukan nilai:

  a = sel kiri atas

  b = sel kanan atas

  c = sel kiri bawah

  d = sel kanan bawah

3. Menghitung total baris pertama:

$$
a+b
$$

4. Menghitung total baris kedua:

$$
c+d
$$

5. Menghitung total kolom pertama:

$$
a+c
$$

6. Menghitung total kolom kedua:

$$
b+d
$$

7. Menghitung total keseluruhan data:

$$
n=a+b+c+d
$$

8. Menghitung faktorial dari seluruh komponen yang diperlukan.

9. Menghitung probabilitas Fisher menggunakan rumus:

$$
P=
\frac
{
(a+b)!(c+d)!(a+c)!(b+d)!
}
{
a!b!c!d!n!
}
$$

10. Membandingkan nilai probabilitas dengan taraf signifikansi α.

11. Jika:

$$
P < \alpha
$$

maka H0 ditolak.

12. Jika:

$$
P \ge \alpha
$$

maka gagal menolak H0.

13. Menarik kesimpulan mengenai hubungan kedua variabel.

### Kelebihan

1. Sangat akurat untuk ukuran sampel kecil.
2. Tidak memerlukan asumsi frekuensi harapan minimum seperti pada uji Chi-Square.
3. Cocok digunakan untuk tabel kontingensi 2×2.
4. Memberikan nilai probabilitas (p-value) yang eksak, bukan pendekatan.
5. Tetap dapat digunakan meskipun terdapat frekuensi observasi yang sangat kecil atau bernilai kurang dari 5.

###Kekurangan
1. Perhitungan manual cukup rumit karena melibatkan operasi faktorial.
2. Kurang efisien untuk ukuran sampel yang sangat besar.
3. Umumnya lebih cocok digunakan pada tabel kontingensi 2×2 sehingga penerapannya lebih terbatas dibanding Chi-Square.
4. Membutuhkan waktu komputasi lebih lama jika jumlah data semakin besar.
5. Tidak memberikan informasi mengenai kuat atau lemahnya hubungan antar variabel, hanya menunjukkan ada atau tidaknya hubungan yang signifikan.

```python
# MENGGUNAKAN LIBRARY
from scipy.stats import fisher_exact

data = [[30, 20], [15, 35]]

odds_ratio, p_value = fisher_exact(data)

alpha = 0.05

if p_value < alpha:
    keputusan = "Tolak H0: terdapat hubungan"
else:
    keputusan = "Gagal menolak H0: tidak terdapat hubungan"

print("data =", data)
print("odds_ratio =", odds_ratio)
print("p_value =", p_value)
print(keputusan)
```

```python
# TANPA LIBRARY STATISTIK
def faktorial(x):
    hasil = 1
    for i in range(1, x + 1):
        hasil *= i
    return hasil


data = [[30, 20], [15, 35]]

a = data[0][0]
b = data[0][1]
c = data[1][0]
d = data[1][1]

row1 = a + b
row2 = c + d

col1 = a + c
col2 = b + d

n = row1 + row2

P = (
    faktorial(row1)
    * faktorial(row2)
    * faktorial(col1)
    * faktorial(col2)
) / (
    faktorial(a)
    * faktorial(b)
    * faktorial(c)
    * faktorial(d)
    * faktorial(n)
)

alpha = 0.05

if P < alpha:
    keputusan = "Tolak H0: terdapat hubungan"
else:
    keputusan = "Gagal menolak H0: tidak terdapat hubungan"

print("P =", P)
print(keputusan)
```

### Kesimpulan

Berdasarkan hasil pengujian Fisher's Exact Test diperoleh nilai probabilitas $(P)$ sebesar 0,0017263216264753742 dengan taraf signifikansi $\alpha = 0,05$. Karena nilai $P < \alpha$ $(0,001726 < 0,05)$, maka $H_0$ ditolak. Dengan demikian, dapat disimpulkan bahwa terdapat hubungan yang signifikan antara kedua variabel kategorik yang diuji. Hasil ini menunjukkan bahwa kedua variabel tidak bersifat independen dan memiliki keterkaitan satu sama lain.
